```python
import numpy as np
import pandas as pd
from sklearn.manifold import MDS
import matplotlib.pyplot as plt
data_cluster=data_groupby.drop(['Spring','Winter','Fall','Exit'],axis=1)
# data_groupby.index
mds = MDS()
a=mds.fit_transform(data_cluster)
# a = mds.embedding_
#plt.scatter(a[0:483,0],a[0:483,1],color='turquoise')
plt.scatter(a[0:483,0],a[0:483,1],color='blue')
# plt.scatter(a[243:483,0],a[243:483,1],color='red')
plt.show()
```
